```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator

#for data visulization and numerical operations 
import matplotlib.pyplot as plt
import numpy as np 

from random import randint
import os
import splitfolders 
import cv2 
from tensorflow.keras.applications.resnet50 import preprocess_input 
from tensorflow.keras import layers, models
```

```python

input_folder = r'D:\MachineLearning\MLDataSets\archive\Agricultural-crops'
output_folder = r'D:\MachineLearning\MLDataSets\ImageRecognition'

#80% training, 10% testing, 10% validation 
split_ratio = (0.8, 0.1, 0.1)

splitfolders.ratio(
    input_folder, 
    output = output_folder, 
    seed = 500,
    ratio = split_ratio, 
    group_prefix=None
)
```

```python
# define parameters 
image_size = (224, 224)
batch_size = 32


train_datagen = ImageDataGenerator(
    # this calls in the pre trained "Resnet" module to each image. The function scales the pixels value to the format our pre trained model uses 
    preprocessing_function = preprocess_input,
    # randomly rotes the image 
    rotation_range = 20,
    # # randomly move image horizontally 
    width_shift_range= 0.2, 
    # randomly move image up and down 
    height_shift_range = 0.2, 
    shear_range = 0.2, 
    zoom_range = 0.2, 
    # randomly flips image horizontally 
    horizontal_flip = True,
    fill_mode = "nearest"
    )
```

```python
# data augmentation for test data (only for rescaling) 
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
# data augmentation for validation data (only for rescaling) 
valid_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
train_dir = os.path.join(output_folder, 'train')
val_dir = os.path.join(output_folder, 'val')
test_dir = os.path.join(output_folder, 'test')

train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical'
)

test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical'
)

valid_data=valid_datagen.flow_from_directory(
    val_dir, 
    target_size=image_size, 
    batch_size=batch_size, 
    class_mode='categorical'
)
```

```python
# get a batch of images and labels
images, labels = next(valid_data)

# select a random iamge from batch
idx = random.randint(0, images.shape[0] - 1)

# display the image 
plt.imshow(images[idx])
plt.show()
```

```python
from keras.applications.resnet import ResNet50 
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(image_size[0], image_size[1],3))

base_model.trainable = False
```

```python
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(30, activation='softmax')
                                                      
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy', 
              metrics=['accuracy']
             )
```

```python
model.fit(train_data, epochs=25, validation_data=valid_data)
```

```python
test_loss, test_accuracy = model.evaluate(test_data)

print(f'Test Accuracy: {test_accuracy * 100:.2f}%')
```
